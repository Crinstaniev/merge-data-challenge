# Part II: Query Ethereum Beacon Chain Block Data

Note 1: you need to add "!" in front of the sample code for it to run successfully in the Colab

Note 2: you need to get an Ethereum node endpoint from a node provider such as quicknode. Please delete "/" at the end of the URL for it to work in the sample code. 

Note 3: you will find a newly generated folder named "output" under the current directory and a new file of data named "beacon_blocks.json" in the output folder. 

```python
!ethereum2etl export_beacon_blocks --start-block 4680000 --end-block 4720000 --output-dir output --output-format json --provider-uri https://ultra-chaotic-silence.discover.quiknode.pro/3a20380951ff287a674246a038a9563feea0e27f
```

```python
block_df = pd.read_json('output/beacon_blocks.json', lines=True)
```

```python
block_df.head()
```
